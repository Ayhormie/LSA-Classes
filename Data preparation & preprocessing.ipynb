```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
my_titanic=pd.read_csv('titanic2019/train.csv')
```

```python
my_titanic
```

```python
my_titanic.isnull()
```

```python
my_titanic.isnull().sum()
```

```python
my_titanic.fillna(-999, inplace=True)
```

```python
my_titanic.drop(['PassengerId', 'Name'],axis=1,inplace=True)
```

```python
my_titanic.info()
```

```python
my_titanic=pd.get_dummies(my_titanic)
```

```python
my_titanic
```

```python
my_titanic.info()
```

```python
y=my_titanic['Survived']
x=my_titanic.drop('Survived',axis=1)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
prediction=model.predict(X_test)
```

```python
prediction
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test, prediction)
```
