```python
year= np.array([1950,
               1970,
               1990,
               2010])
population=np.array([2.519,
                    3.162,
                    5.263,
                    6.972])
```

```python
plt.plot(year,population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Census')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(year,population)
#plt.plot(year,population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Census')

plt.subplot(1,2,2)
plt.scatter(population,year)
# plt.plot(population,year)
plt.xlabel('Population')
plt.ylabel('Year')
plt.title('Census')
plt.tight_layout()
plt.show()
```

```python
#Object Oriented Method
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(year,population)
#Add title and label
ax1.set_xlabel('Population')
ax1.set_ylabel('Year')
ax1.set_title('Census')

ax2=fig.add_axes([0.2,0.4,0.4,0.4])
ax2.plot(population,year)
#Add title and label
ax1.set_xlabel('Year')
ax1.set_ylabel('Population')
ax1.set_title('Census')
```
