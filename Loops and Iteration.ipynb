```python
#Using while loop for even numbers btw 0-100
a=0
while a<=100:
    if a%2==0:
        print(a)
    a=a+1
```

```python
#Using while loop for odd num btw 0-100
a=0
while a<=100:
    a=a+1
    if a%2==0:
        continue
    print(a)
    
    
```
